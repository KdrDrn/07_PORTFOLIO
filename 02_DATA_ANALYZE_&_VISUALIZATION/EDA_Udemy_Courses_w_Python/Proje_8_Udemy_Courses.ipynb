```python
index_vals = df['level'].astype('category').cat.codes
fig = go.Figure(data=go.Splom(
                dimensions=[dict(label='price',
                                 values=df_1['price']),
                            dict(label='num_subscribers',
                                 values=df_1['num_subscribers']),
                            dict(label='num_reviews',
                                 values=df_1['num_reviews']),
                            dict(label='num_lectures',
                                 values=df_1['num_lectures']),
                           dict(label='content_duration',
                                 values=df_1['content_duration'])],
                showupperhalf=False, 
                text=df_1['level'],
                marker=dict(color=index_vals,
                            showscale=False, # colors encode categorical variables
                            line_color='white', line_width=0.5)
                ))


fig.update_layout(
    title='UDEMY Courses',
    width=1000,
    height=1000,
)

fig.show()
```
